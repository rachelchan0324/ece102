```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(-10, 10.2, 0.2)

y = np.exp(-t) * np.cos(2*np.pi * t)

plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Plot of $e^{-t} \cos(2\pi t)$')
plt.grid(True)
display(plt)
```

```python
def relu(t):
  if t < 0:
    return 0
  else:
    return t

y = []
time = np.arange(-5, 5.1, 0.1)
for x in time:
  y.append(relu(x))

plt.plot(time, y)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('ReLU')
plt.grid(True)
display(plt)

```

```python
def even(t, f):
  return 0.5 * (f(t) + f(-t))

def odd(t, f):
  return 0.5 * (f(t) - f(-t))

t = np.arange(-5, 5 + 0.1, 0.1)
y_even = [even(tt, relu) for tt in t]
plt.plot(t, y_even, label = 'even')
y_odd = [odd(tt, relu) for tt in t]
plt.plot(t, y_odd, label = 'odd')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.title("Even and Odd Decomposition of ReLU")
plt.grid(True)
plt.legend()

display(plt)
```
