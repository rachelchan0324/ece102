```python
import numpy as np

def nconv(x, tx, h, th):
  y = np.convolve(x, h) * (th[1] - th[0])
  ty = np.linspace(tx[0] + th[0], tx[-1] + th[-1], len(y))
  return y, ty
```

```python
import matplotlib.pyplot as plt

t = np.linspace(1, 2, 100)
f = 2 * np.ones(len(t))
g = 2 * (t-1) * np.ones(len(t))
y, t = nconv(f, t, g, t)
plt.figure()
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Convolution f(t) * g(t)')
plt.grid(True)
plt.show()
```

```python
t = np.linspace(-1/2, 1/2, 100)
rect = np.ones(len(t))
y, ty = nconv(rect, t, rect, t)
plt.figure()
plt.plot(ty, y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Convolution rect(t) * rect(t)')
plt.grid(True)
plt.show()
```
